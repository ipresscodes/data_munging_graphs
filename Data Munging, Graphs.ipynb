## Visualisation

```python
import matplotlib.pyplot as plt
```

```python
x= [1,2,3]
y= [4,5,2]
plt.plot(x,y)
plt.title('Line Graph')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
from matplotlib import style
style.use('ggplot')
x1=[1,2,3]
x2=[4,5,2]
y1=[7,8,9]
y2=[9,10,7]
plt.plot(x1,y1,'g',label='first',linewidth=5)
plt.plot(x2,y2,'c',label='second',linewidth=5)
plt.title('Style Graph')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.legend()
plt.grid(True,color='k')
plt.show()
```

Bar Graph - for categorical format; qualitative, red blue green, BJP Congress AAP

```python
x1=[1,3,5,7]
y1=[5,2,6,6]
x2=[2,4,6,8]
y2=[6,3,5,2]
plt.bar(x1,y1,label='1st Bar Chart',color='white')
plt.bar(x2,y2,label='2nd Bar Chart',color='pink')
plt.legend()
plt.xlabel('bar number')
plt.ylabel('bar height')
plt.title('Bar Graph')
plt.show()
```

Histrogram - bins are used determine how high the bar should go

```python
sale_values=[12,10,5,7,0,14,15,16,24,10,11,15,30,18,11,10,18,30]
bins=[0,5,10,15,20,25,30]
plt.hist(sale_values,bins,histtype='bar',rwidth=0.6)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Histogram')
plt.show()
```

SCATTER PLOT 

```python
import matplotlib.pyplot as plt

x=[1,3,4,8,5,6,7]
y=[3,5,6,7,8,9,3]

plt.scatter(x,y,label='scatterdots',color='green')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Scatter Plot')
plt.legend()
plt.show()
```

PIE CHART 

```python
labels='Python','C++','Ruby','Java'
sizes=[215,130,245,210]
colors=['m','r','c','k']
explode=(0.1,0,0,0)
plt.pie(sizes,explode,labels,colors,autopct='%1.1f%%',shadow=True,startangle=140)
plt.axis('equal')
plt.show()
```
